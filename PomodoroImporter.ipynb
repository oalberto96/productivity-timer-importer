# Pomodoros

```python
def print_hour(timestamp):
    date = datetime.datetime.fromtimestamp(timestamp)
    return date.strftime("%H:%M")

def print_minutes(timestamp):
    return datetime.datetime.fromtimestamp(timestamp).strftime("%M")



# ['_d', '_dt', '_f', '_id', '_m', '_msc', '_r', '_t', '_ty']
df_pomodoros = json_normalize(data['pomodoros'])
df_pomodoros['_t'] = df_pomodoros['_t'].apply(lambda x: print_hour(x))
df_pomodoros['_d'] = df_pomodoros['_d'].apply(lambda x: print_minutes(x))
df_pomodoros['_m'] = df_pomodoros['_m'].apply(lambda x: print_minutes(x))
df_pomodoros.rename(columns={'_t':'Time', '_d':'Duration', '_dt': 'Date', '_id': 'id'}, inplace=True)
df_pomodoros
```
