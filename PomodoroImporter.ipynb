```python
import json
import datetime
from IPython.display import display, HTML
import pandas as pd
from pandas.io.json import json_normalize

file = open('./Backup_2018-08-08_20-16-53.pctb','r')
data = json.loads(file.read())
```

# Date Utils

```python
date = datetime.datetime.fromtimestamp(20)
# %d is the day number
# %m is the month number
# %b is the month abbreviation
# %y is the year last two digits
# %Y is the all year

def total_time(timestamp):
    hours = timestamp / 60 / 60
    minutes = int((hours % 1) * 60)
    hours = int(hours//1)
    return "{}:{}".format(hours,minutes)
     

def print_datetime(timestamp):
    return datetime.datetime.fromtimestamp(timestamp).strftime("%Y/%m/%d - %H:%M")

def print_hour(timestamp):
    date = datetime.datetime.fromtimestamp(timestamp)
    return date.strftime("%H:%M")

def print_minutes(timestamp):
    return datetime.datetime.fromtimestamp(timestamp).strftime("%M")
```

# Pomodoros

```python
df_pomodoros = json_normalize(data['pomodoros'])

def format_pomodoros(df_original):
    df = df_original.copy()
    df['_t'] = df['_t'].apply(lambda x: print_hour(x))
    df['_d'] = df_pomodoros['_d'].apply(lambda x: print_minutes(x))
    df.rename(columns={'_t':'Time', '_d':'Duration', '_dt': 'Date', '_id': 'Id'}, inplace=True)
    df = df.drop('_msc', 1)
    df = df.drop('_m', 1)
    df = df.drop('_ty', 1)
    df = df.drop('_r', 1)
    df = df.drop('_f', 1)
    return df
format_pomodoros(df_pomodoros)
```

# Work units

```python
new_column_names = {'_d':'Duration', '_dt':'Date', 
                              '_id':'Id', '_p':'Project Id', '_t':'Time'}
df_work_units = json_normalize(data['workUnits'])

def format_work_units(df_original):
    df = df_original.copy()
    df['_d'] = df['_d'].apply(lambda x: print_minutes(x))
    df['_t'] = df['_t'].apply(lambda x: print_hour(x))
    df.rename(columns=new_column_names, inplace=True)
    df= df.drop("_msc", 1)
    df= df.drop("_f", 1)
    df= df.drop("_m", 1)
    return df

format_work_units(df_work_units)
```

# Projects

```python
new_column_names = {'_d':'Duration', '_id': 'Id', '_l':'Last activity',
                    '_n':'Name', '_s':'State', '_p': 'Parent Project',
                   '_t': 'Creation Date', '_w': 'Pomodoro Duration'}
df_projects = json_normalize(data['projects'])

def format_projects(df_original):
    df = df_original.copy()
    df['_d'] = df['_d'].apply(lambda x: total_time(x))
    df['_l'] = df['_l'].apply(lambda x: print_datetime(x))
    df['_t'] = df['_t'].apply(lambda x: print_datetime(x))
    df.rename(columns=new_column_names, inplace=True)
    return df

format_projects(df_projects)

```
